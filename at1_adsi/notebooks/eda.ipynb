## EDA 


In this EDA, we try to answer some questions: - 


- What is the distribution of players across the two outcome classes (TARGET_5Yrs: 1 and 0)? Is the dataset balanced or imbalanced in terms of career lengths?
- What is the correlation between TARGET_5Yrs and other numeric variables such as games played (GP), minutes played (MIN), points per game (PTS), etc.? Does career length have a strong correlation with any particular variable?
- What is the distribution of minutes played (MIN) for players with a career length of 5 years or more (TARGET_5Yrs = 1) compared to players with a shorter career (TARGET_5Yrs = 0)?
- How does the average points per game (PTS) differ between players who have had a career length of 5 years or more (TARGET_5Yrs = 1) and those who have not (TARGET_5Yrs = 0)?
- What is the distribution of offensive rebounds (OREB) for players with a career length of 5 years or more (TARGET_5Yrs = 1) compared to players with a shorter career (TARGET_5Yrs = 0)?
- How does the average number of assists (AST) vary between players who have had a career length of 5 years or more (TARGET_5Yrs = 1) and those who have not (TARGET_5Yrs = 0)?
- Is there a significant difference in steals (STL) between players who have had a career length of 5 years or more (TARGET_5Yrs = 1) and those who have not (TARGET_5Yrs = 0)?
- What is the distribution of blocks (BLK) for players with a career length of 5 years or more (TARGET_5Yrs = 1) compared to players with a shorter career (TARGET_5Yrs = 0)?
- How does MINS played and points PTS impact the target variable ?


```python
# importing libraries

import pandas as pd
import seaborn as sns
```

```python
# reading in the data
data = pd.read_csv('..\\data\\interim\\nba_train_interim.csv')
```

```python
# Question 1 : What is the distribution of players across the two outcome classes (TARGET_5Yrs: 1 and 0)? Is the dataset balanced or imbalanced in terms of career lengths?

# plotting the percentage of players in each class

data['TARGET_5Yrs'].value_counts(normalize = True).plot(kind = 'bar')
```

```python
# We can see that the dataset is imbalanced with more than 80% of the players having a career more than 5 years.
```

```python
# Question 2 : What is the correlation between TARGET_5Yrs and other numeric variables such as games played (GP), minutes played (MIN), points per game (PTS), etc.? Does career length have a strong correlation with any particular variable?

# plotting the correlation matrix for all numeric variables and bigger in size

corr = data.corr()
sns.heatmap(corr, annot = True, cmap = 'Blues').figure.set_size_inches(14, 14)
```

```python
# We can see that not a lot of variables have a strong correlation with the target variable. 
```

```python
# Question 3 : What is the distribution of minutes played (MIN) for players with a career length of 5 years or more (TARGET_5Yrs = 1) compared to players with a shorter career (TARGET_5Yrs = 0)?

# plotting the distribution of minutes played for players with a career length of 5 years or more (TARGET_5Yrs = 1) compared to players with a shorter career (TARGET_5Yrs = 0)

sns.boxplot(x = 'TARGET_5Yrs', y = 'MIN', data = data)
```

```python
sns.histplot(x = 'MIN', hue = 'TARGET_5Yrs', data = data, kde = True)
```

```python
# The players with slightly higher minutes played are more likely to have a career of 5 years or more.
```

```python
# Question 4 : How does the average points per game (PTS) differ between players who have had a career length of 5 years or more (TARGET_5Yrs = 1) and those who have not (TARGET_5Yrs = 0)?

# Plotting the distribution of points per game for players with a career length of 5 years or more (TARGET_5Yrs = 1) compared to players with a shorter career (TARGET_5Yrs = 0)

sns.boxplot(x = 'TARGET_5Yrs', y = 'PTS', data = data)
```

```python
sns.histplot(x = 'PTS', hue = 'TARGET_5Yrs', data = data, kde = True)
```

```python
# Generally players with higher points per game are more likely to have a career of 5 years or more.
```

```python
# Question 5 : What is the distribution of offensive rebounds (OREB) for players with a career length of 5 years or more (TARGET_5Yrs = 1) compared to players with a shorter career (TARGET_5Yrs = 0)?

# Plotting the distribution of offensive rebounds for players with a career length of 5 years or more (TARGET_5Yrs = 1) compared to players with a shorter career (TARGET_5Yrs = 0)

sns.boxplot(x = 'TARGET_5Yrs', y = 'OREB', data = data)
```

```python
# Similar to the previous two questions, players with higher offensive rebounds are more likely to have a career of 5 years or more.
```

```python
# Question 6 : How does the average number of assists (AST) vary between players who have had a career length of 5 years or more (TARGET_5Yrs = 1) and those who have not (TARGET_5Yrs = 0)?

# Plotting the distribution of assists for players with a career length of 5 years or more (TARGET_5Yrs = 1) compared to players with a shorter career (TARGET_5Yrs = 0)

sns.boxplot(x = 'TARGET_5Yrs', y = 'AST', data = data)

```

```python
# Similar to the previous three questions, players with higher assists are more likely to have a career of 5 years or more.
```

```python
# Question 7 : Is there a significant difference in steals (STL) between players who have had a career length of 5 years or more (TARGET_5Yrs = 1) and those who have not (TARGET_5Yrs = 0)?

# Plotting the distribution of steals for players with a career length of 5 years or more (TARGET_5Yrs = 1) compared to players with a shorter career (TARGET_5Yrs = 0)

sns.boxplot(x = 'TARGET_5Yrs', y = 'STL', data = data)
```

```python
sns.histplot(x = 'STL', hue = 'TARGET_5Yrs', data = data, kde = True)
```

```python
# Similar to the previous four questions, players with higher steals are more likely to have a career of 5 years or more.
```

```python
# Question 8 : How does MINS played and points PTS impact the target variable ?

# Plotting a jointplot between MINS and PTS with the target variable

sns.jointplot(x = 'MIN', y = 'PTS', hue = 'TARGET_5Yrs', data = data)
```
